Recursion: A function calling it's self called as recursion.

```python
def fact(n):
    if n == 0:
        return 1
    else:
        return n * fact(n-1)
print(fact(5))
```

Recursion using system functionality of python:

```python
import sys

sys.setrecursionlimit(100)

print(sys.getrecursionlimit())

i = 0

def Hello():
    global i
    i += 1
    print("Hello World")
    Hello()
    
Hello()
```
