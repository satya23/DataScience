```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(15,10))
plt.plot([1, 2, 3, 5], [6, 8, 4, 3])
plt.title("First Chart")
plt.xlabel("X asix")
plt.ylabel("Y asix")
plt.show()
```

```python
plt.plot([1, 2, 3, 5], [6, 8, 4, 3], "go") #go for green circle
plt.title("First Chart")
plt.xlabel("X asix")
plt.ylabel("Y asix")
plt.show()
```

```python
x = np.arange(1, 5)
y = x ** 3
plt.plot([1, 2, 3, 5], [6, 8, 4, 3], "go", x, y, "r^") #go for green circle r^ for red tringle
plt.title("First Chart")
plt.xlabel("X asix")
plt.ylabel("Y asix")
plt.show()
```

```python
plt.subplot(1, 2, 1) # subplot(nrow, ncolumn, index)
plt.plot([1, 2, 3, 5], [6, 8, 4, 3], "go")
plt.title("1st plot")

plt.subplot(1, 2, 2)
plt.plot([2, 4, 5, 3], [1, 2, 2, 5], "r^")
plt.title("2nd plot")

plt.suptitle("Sub plot")
plt.show()
```

```python
plt.subplot(2, 1, 1) # subplot(nrow, ncolumn, index)
plt.plot([1, 2, 3, 5], [6, 8, 4, 3], "go")
plt.title("1st plot")

plt.subplot(2, 1, 2)
plt.plot([2, 4, 5, 3], [1, 2, 2, 5], "r^")
plt.title("2nd plot")

plt.suptitle("Sub plot")
plt.show()
```

```python
x = np.arange(1, 5)
y = x ** 3
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (6, 6))
ax[0, 1].plot([1, 2, 3, 5], [6, 8, 4, 3], "go")
ax[1, 0].plot(x, y, "r^")
ax[0, 1].set_title("Square")
ax[1, 0].set_title("Cubes")
plt.show()
```

```python
divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"] 
div_avg_marks = [67, 45, 78, 89, 94]
plt.bar(divisions, div_avg_marks, color='green')
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Average marks")
plt.show()
```

```python
divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"] 
div_avg_marks = [67, 45, 78, 89, 94]
variance = [2, 5, 3, 7, 9]
# we can pass an argument (with its value)xerr or yerr to depict the variance in our data.
plt.barh(divisions, div_avg_marks,xerr = variance, color = 'green') 
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Average marks")
plt.show()
```

```python
divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"] 
div_avg_marks = [67, 45, 78, 89, 94]
boys_avg_marks = [71, 56, 32, 87, 98]

index = np.arange(5)
width = 0.30

plt.bar(index, div_avg_marks, width, color='green', label = "Division average marks")
plt.bar(index+width, boys_avg_marks, width, color='blue', label = "Boys average marks")

plt.title("Horizontal stacked Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")

plt.xticks(index+width/2, divisions)
plt.legend(loc = 'best') # here legend is best position according to the chart
plt.show()
```

```python
divisions = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"] 
div_avg_marks = [67, 45, 78, 89, 94]
boys_avg_marks = [71, 56, 32, 87, 98]

index = np.arange(5)
width = 0.30

plt.bar(index, div_avg_marks, width, color='green', label = "Division average marks")
plt.bar(index, boys_avg_marks, width, color='blue', label = "Boys average marks", bottom = div_avg_marks)

plt.title("Vertically stacked Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")

plt.xticks(index, divisions)
plt.legend(loc = 'best') # here legend is best position according to the chart
plt.show()
```

```python
firms = ["Firm A", "Firm B", "Firm C", "Firm D"]
market_share = [20, 36, 30, 14]
Explode = [0, 0, 1, 0]
plt.pie(market_share, explode = Explode, labels = firms, shadow = True, startangle = 45)
plt.axis('equal')
plt.legend(title = "List of Firms")
plt.show()
```

```python
x = np.random.randn(1000)
plt.title("Histogram")
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.hist(x, 10)
plt.show()
```

```python
height = np.array([5, 6, 4, 3, 5, 4, 5, 4, 5, 6, 6, 7, 2, 4, 3])
weight = np.array([23, 45, 67, 76, 78, 74, 56, 34, 56, 78, 86, 34, 56, 12, 45])

plt.xlim(20, 90)
plt.ylim(2, 9)
plt.scatter(weight, height)
plt.title("Scatter plot")
plt.ylabel("Height")
plt.xlabel("weight")
plt.show()
```

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection = '3d')
ax.scatter3D(weight, height)
ax.set_xlabel("Weight")
ax.set_ylabel("Height")
plt.show()
```
