Derive Data:

```python
import pandas as pd

air_quality = pd.read_csv("../data/air_quality_no2.csv", index_col=0, parse_dates = True ) 
                                                                            #index_col = 0 to remove the first column
                                                                    # parse_dates = True for making date a readable format
air_quality.head()
```

```python
air_quality["london_mg_percubic"] = air_quality["station_london"] * 1.882 # added a new column london_mg_percubic
air_quality.head()
```

```python
air_quality["ration_paris_antwerp"] = air_quality["station_paris"] / air_quality["station_antwerp"]
air_quality.head()
```

```python
air_quality_renamed = air_quality.rename(
        columns={"station_antwerp": "BETR801",
                 "station_paris": "FR04014",
                 "station_london": "London Westminster"})
air_quality_renamed.head()
```

```python
air_quality_renamed = air_quality_renamed.rename(columns = str.lower)
air_quality_renamed.head()
```
